```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("../data/tmdb_5000_credits.csv")
movies = pd.read_csv("../data/tmdb_5000_movies.csv")
credits.head()
```

```python
movies.head()
```

```python
print(f'Credits dimensions: {credits.shape}')
print(f'Movies data dimensions: {movies.shape}')
```

```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id":"id"})
```

```python
movies_merged = movies.merge(credits_column_renamed, on='id')
movies_merged.head()
```

Preprocessing movie data beginning with data cleaning

```python
movies_cleaned = movies_merged.drop(columns=['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
movies_cleaned.head()
```

```python
print(movies_cleaned.info())
movies_cleaned = movies_cleaned.dropna()
```

```python
print(f"ID: {movies_cleaned.head(1)['id']} | Title: {movies_cleaned.head(1)['original_title']} | Overview {movies_cleaned.head(1)['overview']}")
```

Content based recommendation system - Goal is to receive a movie title as input from the user and recommend moives with similar plots

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode', analyzer='word', ngram_range=(1,3), stop_words='english')
```

```python
tfv_matrix = tfv.fit_transform(movies_cleaned['overview'])
```

Encountered error for rows having NA values in 'overview' column, so removed them and ran Tfidf cells again

```python
for index, row in movies_cleaned.iterrows():
    if pd.isna(row['overview']):
        print(row['id'])
```
